#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
import pytest

def count_vowels(s):
    vowels = "aeiou"
    count = 0

    for char in s.lower():
        if char in vowels:
            count += 1

    return count


def test_count_vowels():
    assert count_vowels("hello") == 2
    assert count_vowels("world") == 1
    assert count_vowels("AEIOU") == 5
    assert count_vowels("xyz") == 0
    assert count_vowels("") == 0
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def is_unique(s):
    return len(s) == len(set(s))

def test_is_unique():
    assert is_unique("abc") == True
    assert is_unique("aabc") == False
    assert is_unique("") == True
    assert is_unique("a") == True
    assert is_unique("abca") == False

test_is_unique()
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def count_bits(n):
    return bin(n).count('1')

def test_count_bits():
    assert count_bits(0) == 0
    assert count_bits(1) == 1
    assert count_bits(5) == 2      # 101
    assert count_bits(7) == 3      # 111
    assert count_bits(8) == 1      # 1000

test_count_bits()
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def prime_factors(n):
    factors = {}
    d = 2

    while d * d <= n:
        while n % d == 0:
            factors[d] = factors.get(d, 0) + 1
            n //= d
        d += 1

    if n > 1:
        factors[n] = factors.get(n, 0) + 1

    result = ""
    for p in sorted(factors):
        if factors[p] == 1:
            result += f"({p})"
        else:
            result += f"({p}**{factors[p]})"

    return result

def test_prime_factors():
    assert prime_factors(86240) == "(2**5)(5)(7**2)(11)"
    assert prime_factors(13) == "(13)"
    assert prime_factors(4) == "(2**2)"
    assert prime_factors(18) == "(2)(3**2)"
    assert prime_factors(1) == ""

test_prime_factors()
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(n):
    total = 0
    k = 0

    while total < n:
        k += 1
        total += k * k

    return k if total == n else "unreal"

def test_pyramid():
    assert pyramid(1) == 1              # 1^2
    assert pyramid(5) == 2              # 1+4
    assert pyramid(14) == 3             # 1+4+9
    assert pyramid(30) == 4             # 1+4+9+16
    assert pyramid(55) == 5
    assert pyramid(2) == "unreal"
    assert pyramid(3) == "unreal"

test_pyramid()
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(n):
    s = str(n)
    length = len(s)

    if length % 2 == 0:
        mid = length // 2
        left = s[:mid - 1]
        right = s[mid + 1:]
    else:
        mid = length // 2
        left = s[:mid]
        right = s[mid + 1:]

    return sum(map(int, left)) == sum(map(int, right))

def test_is_balanced():
    assert is_balanced(1238033) == True   # 1+2+3 = 0+3+3
    assert is_balanced(1234006) == False
    assert is_balanced(56239814) == True
    assert is_balanced(11) == True        # обе стороны пустые → 0 == 0
    assert is_balanced(12321) == True

test_is_balanced()
```
